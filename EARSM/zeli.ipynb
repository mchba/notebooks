
## Zeli model (Zeli et al. 2019 Appendix)

Almost like the model by Lazeroms et al. (2013), but re-written in a more robust and consistent manner.

In summary the model is applied as:

- From last iteration, we have $U$, $V$, $\Theta$, $k$, $k_\theta$ and $\varepsilon$ $\Rightarrow$ $S_{ij}$, $\Omega_{ij}$, $\Gamma_i$ and $\Theta_i$.
- Calculate invariants, $II_S$, $II_\Omega$ and $\gamma^{(\theta)}$.
- Calculate $N$ and $N_\theta$
- Calculate $\beta_1$, $\beta_3$, $\beta_4$, $\lambda_{\theta1}$, $\lambda_{\theta2}$, $\lambda_{\theta3}$, $\lambda_{\theta4}$, $\lambda_{g1}$ and $\lambda_{g3}$. (No need to calculate $\beta_2$, $\lambda_{g2}$ and $\lambda_{g4}$, since they are all 0).
- Calculate $a$ and $\overline{u_i'\theta}$
- From $a_{ij}$ and $k$ we can obtain all the turbulence fluxes, which can be used to solve the mom'm equation. Similarly, the heat fluxes can be used in the temperature equation.
- Finally, the auxillary $k$, $\varepsilon$ and $k_\theta$ equations are solved with the new momentum and temperature solutions.

### Anisotropy

Anisotropy of the 2D Wallin and 2D Zeli models:

$$
a = \beta_1 T_1 + \beta_2 T_2 + \beta_3 T_3 + \beta_4 T_4
$$

Tensors are the same:

$$
\begin{matrix}
 &  Wallin~(2000) & Zeli~(2019) \\
T_1 = &     S & S\\
T_2 =   &     S^2 - \frac{1}{3} II_S \mathbf{I} & S^2 - \frac{1}{3} II_S \mathbf{I}\\
T_3 =   &     \Omega^2 - \frac{1}{3} II_\Omega \mathbf{I} & \Omega^2 - \frac{1}{3} II_\Omega \mathbf{I} \\
T_4 = &    S \Omega - \Omega S &  S \Omega - \Omega S\\
\end{matrix}
$$

Coefficients:

$$
\begin{matrix}
 &  Wallin~(2000) & Zeli~(2019) \\
\beta_1 = &     - \frac{6}{5} \frac{N}{N^2 - 2 II_\Omega} & -\frac{8}{15} \frac{N}{D_u}\\
\beta_2 =   &     0 & 0\\
\beta_3 =   &     0 & -\frac{8}{15} \frac{C_2 C_3 Q_6}{D_u Q_2} \\
\beta_4 = &    - \frac{6}{5} \frac{1}{N^2 - 2 II_\Omega} &  -\frac{8}{15} \frac{C_2}{D_u}\\
\end{matrix}
$$

Beware, that Zeli and Wallin use slightly different definitions of $N$: $N_{Zeli} = \frac{4}{9} N_{Wallin}$ (this is in the non-stratified limit). Constants:

$$
\begin{align} 
C_2 &= \frac{7 c_2 + 1}{11} = \frac{4}{9}\\
C_3 &= 1 - c_3 = 0.7 \\
c_\Theta &= c_{\theta,4} - 1 = -1\\
c_\Gamma &= 2 - c_{\theta g}= 1.25
\end{align}
$$

Helper variables:

$$
\begin{align} 
D_u &\equiv N^2 - 2 C_2^2 II_\Omega\\
Q_2 &\equiv 6 N D_u N_\theta^2 - 2 c_\Theta C_3 \gamma^{(\theta)} N_\theta (6 N^2 + D_u) + 8 c_\Theta^2 C_3^2 N {\gamma^{(\theta)}}^2 \\
Q_6 &\equiv -16 C_3 N c_\Theta^2 {\gamma^{(\theta)}}^2 + 6 c_\Theta N_\theta \gamma^{(\theta)} (5 C_2 D_u + 2 \tilde{D}_u) + 45 c_\Gamma C_2 N_\theta \Gamma^2 D_u \\
\tilde{D}_u &\equiv N^2 + 2 C_2^2 II_\Omega\\
\end{align}
$$

#### Consistency with Wallin (2000) model

In the special case $\frac{d \Theta}{d z} = 0$, we have $\gamma^{(\theta)} = 0$. The temperature variance equation, then becomes:

$$
\frac{D k_\theta}{D t} = - \varepsilon_{\theta} + \mathcal{D}_{k_\theta}
$$

**Not sure if we can extract from this that $k_\theta \rightarrow 0$??** If we can, then $\Gamma_k = 0$ and $\Gamma^2 = 0$. Hence, $Q_6 = 0$, so $\beta_3 \rightarrow 0$.

It is also easy to show that the other $\beta$'s are equal (just remember $N_{Zeli} = \frac{4}{9} N_{Wallin}$). 


In conclusion: The anisotropy of the Zeli model in the non-stratified limit is equal to the Wallin 2D model.



### Heat fluxes

Wikström (2000) and Lazeroms (2013) use the normalized heat flux vector, $\xi_i \equiv \frac{\overline{u_i' \theta}}{\sqrt{k k_\theta}}$, while Zeli (2019) use the dimensional heat flux:

$$
\begin{align}
\overline{u'\theta'} &= \frac{1}{2} (\lambda_{\theta2} + \lambda_{\theta3}) \frac{k^3}{\varepsilon^2} \frac{\partial U}{\partial z} \frac{\partial \Theta}{\partial z} - \lambda_{g3} \frac{k^2 k_\theta}{\varepsilon^2} \frac{g}{2 T_0} \frac{\partial U}{\partial z} \\
\overline{v'\theta'} &= \frac{1}{2} (\lambda_{\theta2} + \lambda_{\theta3}) \frac{k^3}{\varepsilon^2} \frac{\partial V}{\partial z} \frac{\partial \Theta}{\partial z} - \lambda_{g3} \frac{k^2 k_\theta}{\varepsilon^2} \frac{g}{2 T_0} \frac{\partial V}{\partial z} \\
\overline{w'\theta'} &=  (\lambda_{\theta1} - \frac{1}{2} II_\Omega \lambda_{\theta4}) \frac{k^2}{\varepsilon}  \frac{\partial \Theta}{\partial z} - \lambda_{g1} \frac{k k_\theta}{\varepsilon} \frac{g}{T_0} 
\end{align}
$$



### Obtaining $N$ in neutral/unstable

$N \equiv c_1 - 1 + \frac{\mathcal{P} + \mathcal{B}}{\varepsilon}$ (differs from Wallin definiton both because of $\mathcal{B}$ and because of no $4/9$).

In neutral or unstable parrallel shear flow, we have $\gamma^{(\theta)} \geq 0$. $N$ is then given by:

$$
N = -c_1 + 1 + \underbrace{\frac{4}{9} \left( \frac{c_1'}{3} + f_M(P_1,P_2) \right)}_{N^{(S)}} + \underbrace{\frac{c_{\theta1}'}{3} + f_M(P_{b1},P_{b2}) }_{N^{(B)}}
$$

$$
f_M = \begin{cases} f_V\left(R_1 + \sqrt{R_2}\right) + f_V\left(R_1 - \sqrt{R_2}  \right) \quad \quad, R_2 \geq 0 \\ 2 \left( R_1^2  - R_2\right)^{1/6} \mathrm{cos}\left(\frac{1}{3} \mathrm{arccos}\left( \frac{R_1}{R_1^2 - R_2} \right) \right)\quad, R_2 < 0 \end{cases}
$$

$$
f_V(x) = \mathrm{sign}(x) x^{1/3}
$$


$$
\begin{align}
P_1 &= c_1' \left( \frac{1}{27} c_1'^2 + \frac{9}{20} II_S - \frac{2}{3} II_\Omega  \right)\\
P_2 &= P_1^2 - \left( \frac{1}{9} c_1'^2 + \frac{9}{10} II_S + \frac{2}{3} II_\Omega  \right)^3
\end{align}
$$

$$
\begin{align}
P_{b1} &= c_{\theta1}' \left( \frac{1}{27} c_{\theta1}'^2 + \frac{1}{3} \alpha_\theta (c_1 - 1) + \frac{2}{9} c_\Theta \left( 2 C_3 + 1 \right) \gamma^{(\theta)} + \frac{1}{3} c_\Gamma \Gamma^2  \right) - \frac{4}{3} c_\Theta C_3 (c_1 - 1) \gamma^{(\theta)}\\
P_{b2} &= P_{b1}^2 - \left( \frac{1}{9} c_{\theta1}'^2 + \frac{2}{3} \alpha_\theta (c_1 - 1) + \frac{4}{9} c_\Theta \left( 2 C_3 + 1 \right) \gamma^{(\theta)} + \frac{2}{3} c_\Gamma \Gamma^2  \right)^3
\end{align}
$$

$$
\alpha_\theta = c_{\theta1} - \left( c_1 + \frac{1}{r}\right) \frac{1}{2}
$$

### Obtaining $N$ in stable

In stable parrallel shear flow, we have $\gamma^{(\theta)} < 0$. Lazeroms (2013) suggested to obtain $N$ from a two-step procedure.

**Step 1: Solve for $\tilde{N}_0$**

**Step 2: Solve for $N$**

$$
N = c_1 - 1 + II_\Omega \left( \beta_1' + \gamma_{(\theta)} (\beta_{\theta2}' - \beta_{\theta3}') - \Gamma^2 \beta_{g2}' \right) - \left( \gamma^{(\theta)} \left( \lambda_{\theta1}' - \frac{1}{2} II_\Omega \lambda_{\theta4}' \right) + \Gamma^2 \lambda_{g1}' \right)
$$

All of the helper variables with primes, e.g. $\beta_1'$, $\beta_{\theta2}'$, etc., are all evaluated as usually except that $N \rightarrow \tilde{N}_0$ in the expressions.


### Obtaining $N_\theta$

For the smart choice of $c_{\theta5} = 0.5$ (also used by Wikström (2000)), we easily obtain $N_\theta$:

$$
N_\theta = c_{\theta1} + \frac{1}{2} \left( N - c_1 - \frac{1}{r}\right)
$$

, where Zeli (2019) uses $c_{\theta1} = 4.51$, $c_1 = 1.8$ and $r = 0.55$ (Zeli p.56 and p.72). This expression is valid for both stable and unstable.



## The temperature variance equation

For the Zeli model, we also need to solve an equation for $k_\theta$:

$$
\frac{D k_\theta}{Dt} = \underbrace{-\overline{w'\theta'} \frac{d\Theta}{dz}}_{\mathcal{P}_\theta} - \underbrace{\frac{k_\theta \varepsilon}{r k}}_{\varepsilon_\theta} + \underbrace{\frac{d}{dz}\left( \frac{\nu_t}{\sigma_{k_\theta}} \frac{d k_\theta}{dz} \right)}_{D_{k_\theta}}
$$

It is very similar to the $k$-equation, so we re-arrange the equation also:

$$
\frac{D k_\theta}{Dt} - D_{k_\theta} +   \varepsilon_\theta  = \mathcal{P}_\theta
$$

Everything on the LHS is treated as "$A x$", while RHS is "$b$".
