## Rotating plane channel flow

Is a "1D problem" in RANS (only function of y).

- Johnston (1972), experiment
- Launder (1987)
- Gatski and Speziale (1993)
- Andersson (1993), nice explanation of stabilization
- Jongen (1998), describes how it is solved with RANS in 1D using only 200 grid points.
- Wallin (2000)



Notice, some people define $y=0$ at the bottom wall, rotate the channel CW or have the rotation axis at another $y$-position.

The case is often used to test RSM or ARSM against $k$-$\varepsilon$, because the latter is unable to predict the skewed $U$-profile. The figure below is just a sketch, but it is true that the flow will be asymmetric with the $U$-peak towards the suction side and $k$-peak towards pressure side.

<img src="rotating_plane_flow.png">

Properties:

- Steady
- Assumed homogeneous in $x$- and $z$-directions, hence "1D problem"
- Has a pressure profile, $p(y)$. When $\Omega>0$, then $p_{top} < p_{bottom}$, hence "suction side" and "pressure side" 
- Flow peak is pushed towards suction side (may seem counter-intuitive, if one thinks of Coriolis, but I think the explanation is suction)
- Turbulence is dampened at the suction side

To understand these, start at the Navier-Stokes equation with Coriolis and centrifugal forces:

$$
\frac{D \vec{u}}{D t} = - \frac{1}{\rho} \nabla p + \nu \nabla^2 \vec{u} - 2 \vec{\Omega} \times \vec{u} - \vec{\Omega} \times \left( \vec{\Omega} \times \vec{r} \right)
$$

Here $\vec{\Omega} = (0,0,\Omega)$ and $\vec{r}$ is the distance vector from the rotation axis. In 2D flows with the rotation vector perpendicular to the plane, we can use:

$$
- \vec{\Omega} \times \left( \vec{\Omega} \times \vec{r} \right) = \nabla \left( \frac{1}{2} \Omega^2 r^2 \right) 
$$

, which can be combined into an "effective, kinematic pressure" $p_{eff} \equiv \frac{p}{\rho} - \frac{1}{2} \Omega^2 r^2$ to reduce the NS eq.:


$$
\frac{D \vec{u}}{D t} = - \nabla p_{eff} + \nu \nabla^2 \vec{u} - 2 \vec{\Omega} \times \vec{u} 
$$

Takine the Reynolds-average:


$$
\frac{D \vec{U}}{D t} = - \nabla P_{eff} + \nu \nabla^2 \vec{U} - 2 \vec{\Omega} \times \vec{U} - \nabla \cdot \mathbf{\tau} 
$$

Assume steady, unidirectional and fully-developed flow, i.e. $U = U(y)$, then:


$$
0 = - \frac{\partial P_{eff}}{\partial x} + \nu \frac{\partial^2 U}{\partial y^2}   -  \frac{\tau_{12}}{\partial y}  \quad \textrm{(eq. 1)}
$$

$$
0 = - \frac{\partial P_{eff}}{\partial y} - 2 \Omega U   -  \frac{\tau_{22}}{\partial y}  \quad \textrm{(eq. 2)}
$$

Eq. 1 is the same as the channel flow equation, except with $P_{eff}$ instead of $P$. It will make no difference for us, because we fix the pressure term to obtain a specific $u_\tau$. Eq. 2 shows that when $\Omega > 0$ (CCW rotating channel), then there we have $dP_{eff}/dy < 0$! This explains why there is a suction side at the top wall.

The amount of system rotation, $\Omega$, is sometimes expressed with the non-dimensional rotation number, $Ro_\tau \equiv \frac{2 \Omega h}{u_\tau}$. Another non-dimensional number is the "stabilization number" $S \equiv \frac{2 \Omega }{- \frac{dU}{dy}}$, which is the background vorticity (rigid body motion of the channel) divided by flow field vorticity. At points in space where $S > 0$, the turbulence is dampenend, while turbulence is enhanced where $S < 0$ (Anderson 1993, p.164). This is the reason for the skewed TKE profile in the sketch above.

## Flow over flat plate with zero pressure gradient (ZPG)

- Pope (2000, p.298)

<img src="zpg.png">

If the freestream velocity is constant with $x$, then the pressure gradient is zero. Unlike the plane channel flow, the ZPG flow depends on the streamwise position! E.g., in the start of the plate the flow is laminar, while it is fully turbulent further downstream.

## Adverse pressure gradient

Same as ZPG flow over flat-plate, but where the freestream velocity is de-accelerating $\rightarrow$ negative pressure gradient (aka. "adverse" pressure gradient). This setup can be used to investigate airfoil behaviour in an easier way, because of the less complicated geometry.
